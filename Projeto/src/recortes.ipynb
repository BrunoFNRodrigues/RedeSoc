```python
user_final = pd.read_csv('dataframes/user_completo.csv')
```

## Fazendo os recortes

```python
dois_ou_menos_df = user_final[user_final['years'] <= 2]
mais_de_dois_df = user_final[user_final['years'] > 2]
isMultiplayer_df = user_final[user_final['isMultiplayer'] == 1.0]
notMultiplayer_df = user_final[user_final['isMultiplayer'] == 0.0]

dois_menos_multiplayer_df = user_final[(user_final['years'] <= 2) & (user_final['isMultiplayer'] == 1.0)]
dois_menos_notMultiplayer_df = user_final[(user_final['years'] <= 2) & (user_final['isMultiplayer'] == 0.0)]
dois_mais_multiplayer_df = user_final[(user_final['years'] > 2) & (user_final['isMultiplayer'] == 1.0)]
dois_mais_notMultiplayer_df = user_final[(user_final['years'] > 2) & (user_final['isMultiplayer'] == 0.0)]
```

### DOIS OU MENOS ANOS

```python
resultC_menos2 = reg.linear(data=dois_ou_menos_df, formula='clustering ~ total_hours_played + games_played')
resultC_menos2.micro_summary()
```

```python
resultC_menos2.plot_residuals()
```

### MAIS DE DOIS

```python
resultC_mais2 = reg.linear(data=mais_de_dois_df, formula='clustering ~ total_hours_played + games_played')
resultC_mais2.micro_summary()
```

```python
resultC_mais2.plot_residuals()
```
