```python
import pandas as pd
import csv
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

<hr style="border:1px solid gray">

```python
dataset_completo = pd.read_csv("../datasets/steam-200k.csv")
amostra_30k = pd.read_csv("../datasets/steam-200k.csv", nrows=30000)
```

```python
amostra_30k.rename(columns={'151603712':'id'}, inplace=True)
amostra_30k.rename(columns={'1.0':'hours played'}, inplace=True)
amostra_30k_played_only = amostra_30k[amostra_30k.purchase != "purchase"]

dataset_completo.rename(columns={'151603712':'id'}, inplace=True)
dataset_completo.rename(columns={'1.0':'hours played'}, inplace=True)
dataset_completo_played_only = dataset_completo[dataset_completo.purchase != "purchase"]
```

```python
dataset_completo_played_only.head()
```

```python
plt.figure(figsize=(8,5))
dataset_completo_played_only[dataset_completo_played_only["hours played"] < 200]['hours played'].plot(kind='kde', label='Completo')
amostra_30k_played_only[amostra_30k_played_only["hours played"] < 200]['hours played'].plot(kind='kde', label='Amostra 30k')
plt.legend()
plt.title('Distribuição de tempo de jogo (densidade)')
plt.xlabel('Horas de jogo')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,5))
a = dataset_completo_played_only[dataset_completo_played_only["hours played"] < 200]["hours played"]
b = amostra_30k_played_only[amostra_30k_played_only["hours played"] < 200]["hours played"]

plt.hist(a, alpha=0.5, label='Completo', bins=30)
plt.hist(b, alpha=0.5, label='Amostragem 30k', bins=30)
plt.xlabel('Horas de jogo')
plt.ylabel('Ocorrências')
plt.legend()
plt.show()
```
