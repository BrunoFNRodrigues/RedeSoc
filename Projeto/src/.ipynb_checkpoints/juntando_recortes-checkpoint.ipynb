## Carregando DFs

```python
multiplayer_df = pd.read_csv('dataframes/multiplayer.csv')
age_df = pd.read_csv('dataframes/steam_account_age.csv')
user_df = pd.read_csv('dataframes/users_variables.csv')
games_df = pd.read_csv('dataframes/games_variables.csv')
```

```python
user_age_df = pd.merge(user_df, age_df, left_on='id', right_on='user_id')
user_age_df
```

```python
games_multiplayer_df = pd.merge(games_df, multiplayer_df, left_index=True, right_index=True)
games_multiplayer_df
```

```python
#games_multiplayer_df.to_csv('dataframes/games_multiplayer.csv', index=False)
```

```python
amostra_30k = pd.read_csv("../datasets/steam-200k.csv", nrows=30000)
amostra_30k
```

```python
amostra_30k.rename(columns={'151603712':'id'}, inplace=True)
amostra_30k.rename(columns={'1.0':'hours_played'}, inplace=True)
amostra_30k.rename(columns={'The Elder Scrolls V Skyrim':'game'}, inplace=True)
amostra_30k
```

```python
amostra_30k_played_only = amostra_30k[amostra_30k.purchase != "purchase"]
amostra_30k_played_only
```

<hr style="border:2px solid gray">

## Extraindo se usuário joga mais singleplayer ou multiplayer

```python
drop = ['Cópias vendidas', 'Número de jogadores', '_merge', 'Popularidade', 'Temp Med','Jogadores por cópias', 'index' ,'slug', 'Hours Played']
isMultiplayer_df = games_multiplayer_df.drop(drop, axis=1)
```

```python
isMultiplayer_df
```

```python
multiplayer_by_user_df = pd.merge(amostra_30k_played_only, isMultiplayer_df, left_on='game', right_on='Games')
```

```python
multiplayer_by_user_df
```

```python
multiplayer_divided = multiplayer_by_user_df.groupby(['id', 'isMultiplayer'])['hours_played'].sum().to_frame()
multiplayer_divided = multiplayer_divided.reset_index(level=1)
multiplayer_divided = multiplayer_divided.reset_index(level=0)
```

```python
multiplayer_divided 
```

```python
multiplayer_only_max = multiplayer_divided.sort_values('hours_played', ascending=False).drop_duplicates('id').sort_index()
```

```python
multiplayer_only_max
```

```python
multiplayer_only_max = multiplayer_only_max.drop(['hours_played'], axis=1)
multiplayer_only_max
```

```python
user_final = pd.merge(user_age_df, multiplayer_only_max, left_on='id', right_on='id')
```

```python
user_final
```
