## Veterano por quantidades de jogos

```python
g = gt_load('grafos/users_network.net.gz')

from graph_tool import clustering

lc = clustering.local_clustering(g)
    
g.add_vp('clustering', lc)
```

```python
data_hours_played = gt_data(g)
data_hours_played
```

```python
column = "total_hours_played"
data_hours_played[column] = (data_hours_played[column] - data_hours_played[column].min()) / (data_hours_played[column].max() - data_hours_played[column].min())    
```

```python
normalized_hours_played = data_hours_played.copy()
normalized_hours_played
```

```python
normalized_hours_played.index = normalized_hours_played.index.astype(int) 
```

## Tempo total do jogo e número de jogadores com o jogo

---

```python
jogadores_possuem_jogo_df = jogadores_possuem_jogo.to_frame()
jogadores_possuem_jogo_df.rename(columns={'id':'Número de jogadores'}, inplace=True)
jogadores_possuem_jogo_df
```

## Cópias vendidas e quantidade de pessoas que realmente jogaram

```python
amostra_30k.head()
```

```python
amostra_30k_purchased = amostra_30k[amostra_30k.purchase == "purchase"]
amostra_30k_purchased.rename(columns={'The Elder Scrolls V Skyrim':'Games'}, inplace=True)
```

```python
amostra_30k_purchased
```

```python
purchased_only = amostra_30k_purchased.groupby(by = 'Games')['id'].count()
```

```python
purchased_only
```

```python
purchased_only_df = purchased_only.to_frame()
purchased_only_df.rename(columns={'id':'Cópias vendidas'}, inplace=True)
purchased_only_df
```

```python
purchase_play_ratio = pd.merge(purchased_only_df, df_hrs_posessao, how="left", indicator=True, left_index=True, right_index=True)
```

```python
purchase_play_ratio
```

```python
purchase_play_ratio['Hours Played'] = purchase_play_ratio['Hours Played'].fillna(0)
purchase_play_ratio['Número de jogadores'] = purchase_play_ratio['Número de jogadores'].fillna(0)
purchase_play_ratio
```

<hr style="border:1px solid gray">


## Popularidade

```python
purchase_play_ratio['Popularidade'] = (purchase_play_ratio['Cópias vendidas'] / purchase_play_ratio['Cópias vendidas'].sum()) * 100
```

<hr style="border:1px solid gray">


## Tempo médio

```python
purchase_play_ratio['Temp Med'] = (purchase_play_ratio['Hours Played'] / purchase_play_ratio['Número de jogadores'])
```

<hr style="border:1px solid gray">


```python
purchase_play_ratio['Jogadores por cópias'] = (purchase_play_ratio['Número de jogadores'] / purchase_play_ratio['Cópias vendidas'])
```

<hr style="border:1px solid gray">
